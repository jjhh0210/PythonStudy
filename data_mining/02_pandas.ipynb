## 특정 행,열 조회
- df['열'] : 1차원,시리즈로 조회
- df[['열','열2',...]] :2차원, DF로 조회
- df.loc[행조건 , 열 이름(list)] : 특정 조건 만족하는 행들만 조회 가능
    - 마찬가지로 행조건 생략 불가(모든행은 :), 열은 생략 가능
    - 행조건에 그냥 행 인덱스도 들어갈 수 있음
        - df.loc[2] : 인덱스 2인 행 출력
    - **&,| -> 각 조건을 반드시 ()로 묶어줘야**
    - 조건 메소드
        - df[열].**isin**([a,b,..]) : a이거나 b이거나,,, 인 데이터만 조회
        - df[열].**between**(a,b) : a<= x <= b 데이터만 조회(숫자형 data만 가능)
    - 슬라이싱에서 loc 예외)) **a:b 일 때 b도 포함!!!!!**

```python
data['Age'] # 하나의 시리즈 떼어내면 1차원
```

```python
data[['Age']] # 리스트로 시리즈 떼어내면 2차원
```

```python
#조건 : 나이 40~45 살의 생존자 중 1등급석이나 3등급 석이었던 사람들을 나이 내림차순으로 정렬하여 출력
df = data
temp = df.loc[(df['Age'].between(40,45)) & (df['Survived']==1) & (df['Pclass'].isin([1,3])), ['Age','Survived','Pclass']].sort_values('Age',ascending=False)
temp.reset_index(drop=True) #인덱스 다시 부여. drop = false가 디폴트(기존의 인덱스를 index로 출력) ->조회이므로 기존의 temp에는 반영 안됨.
```

```python
#행 인덱스 = 2인 데이터 출력 

#df[2] :오류
print(df.loc[2]) #시리즈 출력
df.loc[[2]] # 데이터프레임 출력
```

```python
df.loc[0:3] #슬라이싱에서 loc 예외)) a:b 일 때 b도 포함!!!!!
```

```python
#특정 열의 값에 대해 집계 가능
df.loc[df['Age'].between(10,19),['Fare','Pclass']].mean()
```

## 집계하기 (groupby, agg)
- 집계함수 : sum, mean, max, min, count
    - 각 열에대한 행의 개수 : df.count()      <---- df.info()로도 확인가능
- df.groupby('~별', as_index = True)[' ~의'].sum() : ~별 ~의 합(집계함수)
    - as_index=True (디폴트) : ~별 이라는 열을 인덱스로 쓸 것.
    - **as_index=False** : 인덱스가 새로 생김. 보통 이걸로 씀 그냥
    - agg([집계함수 이름들]) : 여러 가지 집계를 한번에 수행가능
        - df.groupby( ~ )[ ~ ].agg([max,min,mean])
        - 딕셔너리로 열별 집계함수 지정 가능
            - {'Age':'mean', 'Fare':'sum'}

```python
df.count()
```

```python
df['Fare'].mean()
```

```python
#gropuby

df.groupby(['Sex','Embarked'],as_index=False)['Fare','Age'].mean()
```

```python
df.groupby(['Sex','Embarked'],as_index=False)['Fare','Age'].agg(['max','min','mean'])
```

```python
df.groupby(['Sex','Embarked'],as_index=False)['Fare','Age'].agg({'Age':'mean', 'Fare':'sum'})
```
